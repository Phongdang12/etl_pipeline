```python
import pandas as pd
from datetime import datetime
import json

```

```python
with open('output.json','r',encoding='utf-8') as f:
        response_data = json.load(f)

recall_list = []

for r in response_data:
        recall = {
            'recall_id': r['NID'],
            'title': r['Title'],
            'organization': r['Organization'],
            'product_name': r['Product'],
            'issue': r['Issue'],
            'category': r['Category'],
            'updated_at': r['Last updated']
        }
        recall_list.append(recall)

recall_df = pd.DataFrame(recall_list)
current_timestamp = datetime.now()

recall_df['data_received_at'] = current_timestamp
```

```python
# clean_data
recall_df[['issue_category','category']]=recall_df['category'].str.split('-',n=1,expand=True)
recall_df['updated_at'] = pd.to_datetime(recall_df['updated_at'],format='%Y-%m-%d')
recall_df=recall_df.sort_values(by ='updated_at',ascending=False)


recall_df
```
